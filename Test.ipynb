```python
def happiness_table(year: int, limit: int, ascending: bool = False) -> go.Figure:
    yearly_df = df[df["Year"] == year]
    limited_df = yearly_df.sort_values(
        by = 'happiness_score', 
        ascending = ascending
    )[0:limit]
    
    fig = go.Figure(
        data = [
            go.Table(
                header = dict(
                    values = ['Country', 'happiness_score'], 
                    fill_color = "pink"
                ),
                cells = dict(
                    values = [limited_df['Country'], limited_df['happiness_score']],  
                    fill_color = 'lightgoldenrodyellow'
                )
            )
        ]
    )
    
    fig.update_layout(title = f"Happiness Table {year}")
    
    return fig
```

```python
happiness_table(
    year = 2016,
    limit = 50 
).show()
```

```python
def gdp_table(year: int, limit: int, ascending: bool = False) -> go.Figure:
    yearly_df = df[df["Year"] == year]
    limited_df = yearly_df.sort_values(
        by = 'gdp_per_capita', 
        ascending = ascending
    )[0:limit]
    
    fig = go.Figure(
        data = [
            go.Table(
                header = dict(
                    values = ['Country', 'gdp_per_capita'], 
                    fill_color = "pink"
                ),
                cells = dict(
                    values = [limited_df['Country'], limited_df['gdp_per_capita']],  
                    fill_color = 'lightgoldenrodyellow'
                )
            )
        ]
    )
    
    fig.update_layout(title = f"GDP per capita Table {year}")
    
    return fig
```

```python
gdp_table(2015, 50).show()
```

```python
def generosity_table(year: int, limit: int, ascending: bool = False) -> go.Figure:
    yearly_df = df[df["Year"] == year]
    limited_df = yearly_df.sort_values(
        by = 'generosity', 
        ascending = ascending
    )[0:limit]
    
    fig = go.Figure(
        data = [
            go.Table(
                header = dict(
                    values = ['Country', 'generosity'], 
                    fill_color = "pink"
                ),
                cells = dict(
                    values = [limited_df['Country'], limited_df['generosity']],  
                    fill_color = 'lightgoldenrodyellow'
                )
            )
        ]
    )
    
    fig.update_layout(title = f"generosity Table {year}")
    
    return fig
```

```python
generosity_table(2015, 50).show()
```

```python
def GlobalMap(color: str):
    fig = px.choropleth(
        df,
        locations = 'Country',
        locationmode = 'country names',
        # projection = 'orthographic',
        color = color,
        hover_name = 'Country',
        color_continuous_scale = px.colors.sequential.Plasma
    )
    
    return fig
```
